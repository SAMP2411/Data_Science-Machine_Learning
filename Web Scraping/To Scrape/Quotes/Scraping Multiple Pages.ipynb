```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from tqdm import tqdm
link ='https://quotes.toscrape.com/'
# print(requests.get(link).text)
res= requests.get(link)
# print(res.text)
soup = BeautifulSoup(res.text, 'html.parser')
if soup.find('span'):
    print(True)
else:
    print(False)
# type(soup)

# print(soup.find('span', class_ = 'text').text[1:-1])
```

```python
data = []


for sp in soup.find_all('div',class_="quote"):
    # print(sp)
    quote = sp.find('span', class_ = 'text').text[1:-1]
    author = sp.find('small', class_ = 'author').text
    author_id = sp.find('a').get('href')
    tags_list = [  ]
    for tag  in sp.find_all ('a',class_ = 'tag'):
        tags_list.append(tag.text)

    tags = ','.join(tags_list)
    author_link = link + author_id[1:]

    data.append([quote,author,author_id,tags,author_link])
    
```

```python
data = [ ]

for page in tqdm(range(2,11)):
    newlink = link +"page/"+str(page)
    res = requests.get(newlink)
    
    soup = BeautifulSoup(res.text, 'html.parser')

    for sp in soup.find_all('div',class_="quote"):
        # print(sp)
        quote = sp.find('span', class_ = 'text').text[1:-1]
        author = sp.find('small', class_ = 'author').text
        author_id = sp.find('a').get('href')
        tags_list = [  ]
        for tag  in sp.find_all ('a',class_ = 'tag'):
            tags_list.append(tag.text)

        tags = ','.join(tags_list)
        author_link = link + author_id[1:]

        data.append([quote,author,author_id,tags,author_link])

df  =pd.DataFrame(data, columns = ["quote", "author", "author_id", "tags","author_link"] )
df.to_csv('Quotes.csv',index = False)
```
