```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.axes as ax
import pandas as pd

from sklearn import datasets
# import warnings
# warnings.filterwarnings("ignore")
```

```python
data = datasets.load_diabetes()
```

```python
x = data.data
```

```python
x
```

```python
y = data.target
```

```python
features = data.feature_names
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(100)
rf = rf.fit(x_train, y_train)

prediction = rf.predict(x_test)
```

```python
prediction.shape
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, prediction)
```

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=5, shuffle=False)

for train_index, test_index in kfold.split(x):
    print("Train - \n", train_index, "\nTest - \n", test_index)
    print("==========\n")
# kfold.split(x)
```

```python
rf1 = RandomForestRegressor()
error = []

n_estimators = [100,200,500]

for n_est in n_estimators:
    for train_index, test_index in kfold.split(x):
        x_train, x_test = x[train_index], x[test_index]
        y_train, y_test = y[train_index], y[test_index]

        rf1 = RandomForestRegressor(100)

        rf1 = rf1.fit(x_train, y_train)

        prediction = rf1.predict(x_test)
        error.append(mean_squared_error(y_test, prediction))
    
    print("Results for n_estimator : ", n_est)
    print(error)
    print("Average error = ", sum(error)/len(error))
```
